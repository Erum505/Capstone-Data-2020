```python
import pandas as pd

df = pd.read_csv("capstone_dataset_fall_2020_1.csv")
df
```

```python
iceland_df = df[df["CountryName"] == "Iceland"]

#cleaning date data
import datetime
from datetime import timedelta
x = datetime.datetime(2020, 1, 1)

dates = []
for y in range(0,258):
    x = x + timedelta(days=1)
    dates.append(x)
    

len(dates)
dates[10]
```

```python
import matplotlib.pyplot as plt


plt.plot(dates, iceland_df["GovernmentResponseIndex"])
plt.ylabel('Government Response Index Iceland')
plt.xlabel('Date')
plt.suptitle('COVID Government Response Index of Iceland Over Time')

plt.show()

```
